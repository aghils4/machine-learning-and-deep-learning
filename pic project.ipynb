```python
import cv2
import mahotas
import os
import numpy as np
import pandas as pd
```

```python
tr_data=os.listdir('train/')
te_data=os.listdir('test/')
tr_data
```

```python
x_test=[]
```

```python
x_train=[]
y_train=[]
file = os.listdir(os.path.join(os.path.join('train','positive')))
for x in file:
    image=cv2.imread(os.path.join(os.path.join('train','positive'),x))
    image=cv2.resize(image,(500,500))
    image_hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
    hist=cv2.calcHist([image],[0,1,2],None,[8,8,8],[0,256,0,256,0,256])
    cv2.normalize(hist,hist)
    histogram_feattuers=hist.flatten()
    image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    hun_features=cv2.HuMoments(cv2.moments(image_gray)).flatten()
    haralick_features=mahotas.features.haralick(image_gray).mean(axis=0)
    current_features=np.hstack([histogram_feattuers,hun_features,haralick_features])
    x_train.append(current_features)
    y_train.append('1')
file1 = os.listdir(os.path.join(os.path.join('train','negative')))
for y in file1:
    image1=cv2.imread(os.path.join(os.path.join('train','negative'),y))
    image1=cv2.resize(image1,(500,500))
    image_hsv1=cv2.cvtColor(image1,cv2.COLOR_BGR2HSV)
    hist1=cv2.calcHist([image1],[0,1,2],None,[8,8,8],[0,256,0,256,0,256])
    cv2.normalize(hist1,hist1)
    histogram_feattuers1=hist1.flatten()
    image_gray1=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
    hun_features1=cv2.HuMoments(cv2.moments(image_gray1)).flatten()
    haralick_features1=mahotas.features.haralick(image_gray1).mean(axis=0)
    current_features1=np.hstack([histogram_feattuers1,hun_features1,haralick_features1])
    x_train.append(current_features1)
    y_train.append('0')
```

```python
pd.DataFrame(x_train)
```

```python
file2 = os.listdir(os.path.join(os.path.join('test')))
for z in file2:
    image2=cv2.imread(os.path.join(os.path.join('test'),z))
    image2=cv2.resize(image2,(500,500))
    image_hsv2=cv2.cvtColor(image2,cv2.COLOR_BGR2HSV)
    hist2=cv2.calcHist([image2],[0,1,2],None,[8,8,8],[0,256,0,256,0,256])
    cv2.normalize(hist2,hist2)
    histogram_feattuers2=hist2.flatten()
    image_gray2=cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)
    hun_features2=cv2.HuMoments(cv2.moments(image_gray2)).flatten()
    haralick_features2=mahotas.features.haralick(image_gray2).mean(axis=0)
    current_features2=np.hstack([histogram_feattuers2,hun_features2,haralick_features2])
    x_test.append(current_features2)
```

```python
pd.DataFrame(x_test)
```

```python
pd.DataFrame(y_train).value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
X_train , X_test , Y_train , Y_test=train_test_split(x_train,y_train,test_size=0.0001,random_state=44)
len(X_train)
```

```python
params={'n_estimators':[2000],'max_depth':[15],'min_samples_leaf':[20]}
```

```python
model=RandomForestClassifier()
```

```python
clf=GridSearchCV(model,param_grid=params,cv=3,verbose=100)
```

```python
clf.fit(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
clf.score(X_train,Y_train)
```
